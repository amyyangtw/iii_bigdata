## 簡體轉繁體

```python
from hanziconv import HanziConv
zhwiki_file = open('zhwiki.txt', 'r')
with open('zhwiki_cht.txt', 'w') as cht:
    for line in iter(zhwiki_file):
        #print(line)
        #print(HanziConv.toTraditional(line))
        cht.write(HanziConv.toTraditional(line)+'\n')
    zhwiki_file.close()
```

### 載入繁體中文詞庫

```python
#print(jieba.__file__)
dict_txt = open('dict.txt', 'r')
with open('dict.txt.big', 'w') as cht:
    for line in iter(dict_txt):
        cht.write(HanziConv.toTraditional(line))
    dict_txt.close()
jieba.set_dictionary('dict.txt.big')
```

```python
# 去除停用詞
stopword_set = set()
with open('stopwords.txt','r', encoding='utf-8') as stopwords:
    for stopword in stopwords:
        stopword_set.add(stopword.strip('\n'))
```

# 斷字詞

```python
import jieba
import logging

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

# 使用繁體詞庫，對繁體中文的斷詞會比較精確
jieba.set_dictionary('dict.txt.big')

jieba.add_word('滷肉飯')

output = open('wiki_seg.txt', 'w', encoding='utf-8')
with open('zhwiki_cht.txt', 'r', encoding='utf-8') as content :
    for texts_num, line in enumerate(content):
        line = line.strip('\n')
        words = jieba.cut(line, cut_all=False)
        for word in words:
            if word not in stopword_set:
                output.write(word + ' ')
        output.write('\n')

        if (texts_num + 1) % 10000 == 0:
            logging.info("已完成前 %d 行的斷詞" % (texts_num + 1))
output.close()
```

# Step 5 訓練模型


```python
# 可以練習調整model的各種參數

import logging
from gensim.models import word2vec

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
sentences = word2vec.LineSentence("wiki_seg.txt")
model = word2vec.Word2Vec(sentences, size=250)

#保存模型，供日後使用
model.save("word2vec.model")
```

# Step 6 測試模型

```python
model = word2vec.Word2Vec.load("word2vec.model")
model.wv.most_similar("滷肉飯")
```

```python
from gensim.models import word2vec
from gensim import models
import logging

def main():
    logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

    print("提供 3 種測試模式\n")
    print("輸入一個詞，則去尋找前一百個該詞的相似詞")
    print("輸入兩個詞，則去計算兩個詞的餘弦相似度")
    print("輸入三個詞，進行類比推理")

    while True:
        try:
            query = input()
            q_list = query.split()

            if len(q_list) == 1:
                print("相似詞前 10 排序")
                res = model.most_similar(q_list[0],topn = 10)
                for item in res:
                    print(item[0]+","+str(item[1]))

            elif len(q_list) == 2:
                print("計算 Cosine 相似度")
                res = model.similarity(q_list[0],q_list[1])
                print(res)
            else:
                print("%s之於%s，如%s之於" % (q_list[0],q_list[2],q_list[1]))
                res = model.most_similar([q_list[0],q_list[1]], [q_list[2]], topn= 10)
                for item in res:
                    print(item[0]+","+str(item[1]))
            print("----------------------------")
        except Exception as e:
            print(repr(e))
            break

if __name__ == "__main__":
    main()
```
